**AdaGrad Optimizer (Adaptive Gradient Optimizer)**

Adagrad is an optimizer with parameter-specific learning rates, which are adaptive relative to how frequently a parameter gets updated during training.The more update a parameter recieves the smaller the learning rate.

**Two Features in learning stage in NN**

Dense means a data representation in the form of Matrix whose values are Not Zero.

Sparse means a data representation in the form of Matrix whose values are Zero.
